## Does Economic Inequality Predict Happiness? 
#### A Data‑Driven Exploration of Global Well‑Being

### Introduction
Happiness is something everyone cares about, yet it often feels mysterious. Some countries seem to consistently report higher life satisfaction, while others struggle year after year. It’s easy to assume that wealth alone explains the difference, but the story is rarely that simple. In recent years, more attention has been drawn to the idea that economic inequality — the gap between the richest and poorest people — might play a role in how happy a society feels.

The idea sounds intuitive. When opportunities are unevenly distributed, trust can be weakened, stress can rise, and people may feel that life is less fair. But is this actually reflected in the data? Or is inequality just one of many factors that shape well‑being?

To explore this question, two large and respected sources of real‑world data are used in this project. The World Happiness Report (2005–2024) provides yearly measurements of how satisfied people feel with their lives, along with indicators like social support, freedom, and health. The World Bank’s Development Indicators dataset offers the Gini Index, one of the most widely used measures of economic inequality, covering many countries and decades.

By bringing these two datasets together, a simple but meaningful question can be explored:

Does economic inequality predict how happy a country is?
The goal of this project is not only to look for statistical relationships, but also to tell a clear and accessible story about what the data suggests. Patterns across regions will be examined, trends over time will be visualized, and several hypotheses will be tested to see whether inequality truly stands out as a key factor in national well‑being.

In the end, the hope is that the findings will offer a clearer picture of how fairness, opportunity, and economic balance might shape the emotional landscape of societies around the world — and why this matters for all of us.

### Problem Formulation
Happiness is often treated as something personal and subjective, yet it is measured every year on a global scale. Countries are ranked, compared, and analyzed based on how satisfied their citizens feel with their lives. At the same time, economic inequality continues to be discussed as one of the major forces shaping social well‑being. The idea that large gaps between the rich and the poor might influence how people feel about their lives has been suggested many times, but the strength of this relationship is still debated.

In this project, the focus is placed on exploring whether a connection between economic inequality and national happiness can be observed in real‑world data. Happiness is represented through the Ladder Score from the World Happiness Report, which reflects how people rate their overall life satisfaction. Inequality is represented through the Gini Index, a widely used measure provided by the World Bank, where higher values indicate greater inequality.

By combining these two independent datasets, a clearer picture can be formed of how inequality and happiness move together across countries and over time. The goal is not to prove causation, but to understand whether meaningful patterns appear when the two measures are compared.

Several guiding questions shape this investigation:

- Do countries with higher inequality tend to report lower happiness?

- Are certain regions more affected by inequality than others?

- Does economic prosperity (such as GDP per capita) change the relationship between inequality and happiness?

- How stable is this relationship over time?

Through data cleaning, visualization, and statistical testing, the project aims to translate a broad social question into a structured, data‑driven analysis. The hope is that the findings will offer a clearer understanding of how fairness and opportunity might influence the emotional well‑being of entire societies.
